## NumPy
We'll now import NumPy. NumPy provides high-performance multidimensional arrays and linear algebra operations (similar to Matlab). It is a fundamental package for scientific computing with Python.

More information: https://numpy.org/doc/1.18/user/index.html

```python
import numpy as np
```

```python
# NumPy arrays
a = np.array(a) # a is now a NumPy array
b = a*2

print(b)
```

```python
c = np.array([[45,36,18,12],[53,90,56,62],[10,84,27,49]])
```

```python
c
```

```python
# Indexing
print(c[0, :])  # First row
print(c[:, 0])  # First column
print(c[1, 1])  # Second row, second column
print(c[1:3, :])  # Second rows
```

### Arithmetical operations in Numpy

Information: https://docs.scipy.org/doc/numpy-1.13.0/reference/routines.math.html


```python
# Get the minimum horizontally across columns (axis 1)
c.sum(axis=0)
```

```python
print(c.sum(axis=1))  # sum across columns
print(c.sum())  # sum entire array
```

```python
# Get the minimum horizontally across columns (axis 1)
c.max(axis=1)
```

```python
# Boolean operations
c > 40
```

```python
c
```

```python
c[c > 40]
```

## Masked arrays

Information: https://docs.scipy.org/doc/numpy/reference/maskedarray.generic.html

Useful in working with no-data values like -9999


```python
import numpy.ma as ma
```

```python
d = ma.masked_greater(c, 40)
print(d)
```

## Pandas
A library to work with tabular data

Comparable to data frames in R

Pros:
You can name columns and refence by labels instead of numeric indices like in NumPy arrays
This also makes performing group operations easier and more readable

Cons:
Designed for tabular data, not for dimensions > 2. We will cover xarray for N-dimensional data later.

More information: https://pandas.pydata.org/docs/user_guide/index.html
